## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df["Subtotal"] = df["qty"] * df["unit_price"]
df
```

```python
# Assuming df is already defined and contains the required columns
# pound for items 50 pounds or under.

def shipping_price(x):
    if x["unit_weight"] * x["qty"] > 50:
        return (7 * (x["unit_weight"]) * x["qty"])
    else:
        return (10 * x["unit_weight"]) * x["qty"]

df["Shipping Price"] = df.apply(shipping_price, axis=1).round(2)
df["Shipping Price"].sort_values(ascending=True)
#df
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
def total_price(x):
    return ((x["Subtotal"] + x["Shipping Price"]) * 0.0925) + x["Subtotal"] + x["Shipping Price"]
df["Total Price"] = df.apply(total_price, axis=1).round(2)
df
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
def subtotalandshipping(x):
    return (x["unit_cost"] * x["qty"] + x["Shipping Price"])
df["Total Cost"] = df.apply(subtotalandshipping, axis=1).round(2)

df
```

```python
# Create a column for the profit of
# each line using line cost and line price
def profitfunc(x):
    return (x["Total Price"] - x["Total Cost"])

df["Profit"] = df.apply(profitfunc, axis=1).round(2)
df
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
#best_clients_list
df.set_index(df["client_id"], inplace=True)
total_best_client = df.loc[best_clients_list, "Total Price"].groupby("client_id").sum()
total_best_client
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
top_clients_df = df.loc[best_clients_list]
summary_data = top_clients_df.groupby(top_clients_df.index).agg(
    total_units_purchased=pd.NamedAgg(column='qty', aggfunc='sum'),
    total_shipping_price=pd.NamedAgg(column='Shipping Price', aggfunc='sum'),
    total_revenue=pd.NamedAgg(column='Total Price', aggfunc='sum'),
    total_profit=pd.NamedAgg(column='Profit', aggfunc='sum')
)
summary_sorted_by_profit = summary_data.sort_values('total_profit', ascending=False)
summary_sorted_by_profit
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
df.rename(columns={
    'qty': 'Total Units Purchased',
    'Shipping Price': 'Total Shipping Price ($M)',
    'Profit': 'Total Profit ($M)'
}, inplace=True)

df['Total Shipping Price ($M)'] = (df['Total Shipping Price ($M)'] / 1_000_000).round(2)
df['Total Profit ($M)'] = (df['Total Profit ($M)'] / 1_000_000).round(2)
df
```
